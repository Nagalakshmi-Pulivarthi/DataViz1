### Student Performance Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
student_data=pd.read_csv("Student_performance_data _.csv")
```

## Converting raw data into a readable format

```python
student_df =student_data.copy()
student_df.head()
student_df['Gender'] = student_df['Gender'].replace({0:'male', 1:'female'})
student_df['Ethnicity'] = student_df['Ethnicity'].replace({0: "Caucasian", 
                                           1: "African American",
                                           2: "Asian",
                                           3: "Other"})
student_df['ParentalEducation'] = student_df['ParentalEducation'].replace({0: "None",
                                                              1: "High School",
                                                              2: "Some College",
                                                              3: "Bachelor's",
                                                              4: "Higher"})
student_df['ParentalSupport'] = student_df['ParentalSupport'].replace({0: "None",
                                                              1: "Low",
                                                              2: "Moderate",
                                                              3: "High",
                                                              4: "Very High"})
student_df['Tutoring'] = student_df['Tutoring'].replace({0: "No", 1 : "Yes"})
student_df['GradeClass'] =student_df['GradeClass'].replace({0: "A", 1: "B", 2: "C", 3: "D", 4: "F"})
student_df.head()
```

```python
student_df.head()
```

## Group of Pie charts for quick visulization 

```python
df_fill =student_df[['Gender', 'Ethnicity', 'ParentalEducation', 'ParentalSupport', 'Tutoring', 'GradeClass']]

for column in df_fill:
  plt.figure(figsize=(4, 4))
  df_fill[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
  plt.title(column)
  plt.ylabel('')
  plt.show()
```

```python
numerical = student_df.select_dtypes(include=['float64'])
numerical.head()
```

```python
new_data = df_fill.join(numerical)
new_data.head()
```

## GPA Distribution by Ethinicity

```python
average_gpa_per_ethnicity = new_data.groupby('Ethnicity')['GPA'].mean().reset_index()
print(average_gpa_per_ethnicity)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Ethnicity', y='GPA', data=new_data)
plt.title('GPA Distribution by Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('GPA')
plt.show()
```

### GPA Distribution on Ages

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='GPA', data=student_data)
plt.title('GPA Distribution across Age Groups')
plt.xlabel('Age')
plt.ylabel('GPA')
plt.show()
```

### GPA Distribution by Parental Education

```python
average_gpa_per_ParentalEducation = new_data.groupby('ParentalEducation')['GPA'].mean().reset_index()
print(average_gpa_per_ParentalEducation)
plt.figure(figsize=(10, 6))
sns.violinplot(x='ParentalEducation', y='GPA', data=new_data)
plt.title('GPA Distribution by Parental Education')
plt.xlabel('Parental Education')
plt.ylabel('Score GPA')
```

### How all the factors are corralted to the GPA

```python
students_correlation= student_data.corr()
plt.figure(figsize=(10,10))
axis_corr = sns.heatmap(students_correlation,annot=True, cmap='crest', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

## Impacts of extracurricular activities on GPA

```python
plt.figure(figsize=(10, 6))
sns.pairplot(student_data, vars=['GPA', 'Sports', 'Music', 'Volunteering'])
plt.suptitle('Relationship between Extracurricular Activities and GPA', y=1.02)
plt.show()

sports_gpa_corr = student_data['Sports'].corr(student_data['GPA'])
music_gpa_corr = student_data['Music'].corr(student_data['GPA'])
volunteering_gpa_corr = student_data['Volunteering'].corr(student_data['GPA'])
print(f'Sports-GPA correlation: {sports_gpa_corr:.2f}')
print(f'Music-GPA correlation: {music_gpa_corr:.2f}')
print(f'Volunteering-GPA correlation: {volunteering_gpa_corr:.2f}')
```
